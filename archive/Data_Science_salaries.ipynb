# Understanding the data

The dataframe has 11 columns (without unnamed): 
1. Work year: The year that data was collected 
2. Experience_level: The experience_level in the job, or you can understand that experience level is your level in companies, including:
* EN = Entry level
* MI = Mid level
* SE = Senior level
* EX = Executive level
3. Employment_type: Employment_type helps you to know whether the staff works full-time or part-time.... This includes:
* PT = Part-time
* FT = Full-time
* CT = Contract
* FL = Freelance
4. Job_title: Job_title is the name of your job
5. Salary: the money that staff is paid due to their work
6. Salary_currency: The currency that salary is paid. It can be paid in USD, EURO,...
7. SalaryinUSD: The salary is paid in USD (United State Dollar)
8. Employee_residence: Employee_residence is the nation that staff was working in
9. remote_ratio: this calculate the amount of work done remotely:
* 0 = no remote work (work onsite)
* 50 = Partially remote (half of work is done remotely, half of work is done onsite)
* 100 = fully remote
10. company_loacation: Determining the nation of your company
11. Company_size: the size of companies that staff has worked in, including:
* S = less than 50 employees (small)
* M = 50 to 250 employees (medium)
* L = more than 250 employees (large)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
X=df5.drop('salary_in_usd',axis=1)
y=df5['salary_in_usd']
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=10)
pipeline= Pipeline([('one_hot_cat',OneHotEncoder(handle_unknown='ignore'))])
X_train= pipeline.fit_transform(X_train)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv= ShuffleSplit(n_splits= 5,test_size=0.2, random_state=10)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
```

```python
def find_best_model_using_GridsearchCV(X,y):
    algos= {'linear Regression':{'model':LinearRegression(),'params':{'normalize':[True,False]}},'lasso':{'model':Lasso(),'params':{'alpha':[1,2],'selection':['random','cyclic']}},'decision tree':{'model':DecisionTreeRegressor(),'params':{'criterion':['mse','friedman_mse'],'splitter': ['best','random']}},'ridge':{'model':Ridge(),'params':{'alpha':[1,2],'fit_intercept':[True,False],'normalize':[True,False],'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs']}},'svc':{'model':SVC(),'params':{'C':[1.0,2.0],'kernel':['linear', 'poly', 'rbf'],'degree':[1,2,3],'gamma':['scale', 'auto']}}}
    scores=[]
    cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
    for algo_name, config in algos.items():
        gs= GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
        gs.fit(X,y)
        scores.append({'model':algo_name,'best_score':gs.best_score_,'best_params':gs.best_params_})
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
     
    
```

```python
find_best_model_using_GridsearchCV(X_train,y_train)
```
